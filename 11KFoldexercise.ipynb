```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data)
df['target'] = iris.target
df['flower_names'] = df.target.apply(lambda x: iris.target_names[x])
df
```

```python
from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits = 3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
scores_lr = []
scores_svm = []
scores_rfc = []
```

```python
def get_scores(model, X_train, X_test, Y_train, Y_test):
  model.fit(X_train, Y_train)
  return model.score(X_test, Y_test)
```

```python
for train_index, test_index in folds.split(iris.data, iris.target):
  X_train, X_test, Y_train, Y_test = iris.data[train_index], iris.data[test_index], iris.target[train_index], iris.target[test_index]
  scores_lr.append(get_scores(LogisticRegression(max_iter=1000), X_train, X_test, Y_train, Y_test))
  scores_svm.append(get_scores(SVC(), X_train, X_test, Y_train, Y_test))
  scores_rfc.append(get_scores(RandomForestClassifier(), X_train, X_test, Y_train, Y_test))
```

```python
np.mean(scores_lr)
```

```python
np.mean(scores_svm)
```

```python
np.mean(scores_rfc)
```
