```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("Homeprices.csv")
df
```

```python
import math
```

```python
df.bedroom = df.bedroom.fillna(math.floor(df.bedroom.median()))
df
```

```python
model = linear_model.LinearRegression()

model.fit(df[['area', 'bedroom', 'age']], df.price)
```

```python
model.predict([[3000, 3, 10]])
```

```python
import pickle
with open('model_pickle', 'wb') as f:
  pickle.dump(model, f)
```

```python
with open('model_pickle', 'rb') as f:
  model_read = pickle.load(f)
```

```python
model_read.predict([[3000, 3, 10]])
```
